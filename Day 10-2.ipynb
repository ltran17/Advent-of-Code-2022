```python
def get_lines(file_type='sample'):
    '''
    Read in the lines of today's sample/input file line by line. 
    Assumes the file is in folder called 'inputs/'
    
    Parameters
    ----------
    file_type : str
        Either sample or input
    
    Returns
    -------
    list of inputs stripped of whitespace
    '''
    import datetime
    day = str(datetime.datetime.today().day).zfill(2)
    filename = f'inputs/{day}-{file_type}.txt'
    try:
        with open(filename,'r') as file:
            lines = [line.strip() for line in file.readlines()]
        return lines
    except:
        print(filename+' does not exist')
```

```python
def initialize(start_value):
    x = [start_value]
    return x
```

```python
def noop(x):
    current = x[-1]
    x.append(current)
    return x
```

```python
def addx(x, val):
    current = x[-1]
    x.append(current)
    x.append(current+val)
    return x
```

```python
def get_sprite_positions(x, instructions):
    for instruction in instructions:
        if instruction.split(' ')[0] == 'noop':
            x = noop(x)
        else:
            value = int(instruction.split(' ')[1])
            x = addx(x, value)
    return x
```

```python
instructions = get_lines('sample')
x = initialize(1)
x = get_sprite_positions(x,instructions)
```

```python
instructions = get_lines('input')
x = initialize(1)
x = get_sprite_positions(x,instructions)
cycles = len(x)
for cycle in range(cycles-1):
    row = cycle%40
    if row-1 <= x[cycle] <= row+1:
        print('# ', end='')
    else:
        print('  ', end='')
    if (cycle+1)%40 == 0:
        print('\n')
```
